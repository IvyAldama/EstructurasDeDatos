```python
import pandas as pd
from sklearn import datasets
```

```python
iris_sk = datasets.load_iris(as_frame=True)
iris = pd.DataFrame(data=iris_sk.data, columns=iris_sk.feature_names)

# print(iris_sk.target)
iris['target'] = pd.Series(iris_sk.target)
target_names = pd.Series(iris_sk.target_names)
```

```python
print(iris)
print(target_names)
```

```python
iris.describe()
```

```python
# normalization
columns = list(iris.columns)
columns.remove('target')
print(columns)


for  col in columns :
  col_min = iris[col].min()
  col_max = iris[col].max()
  div = col_max-col_min

  iris[col] = (iris[col] -  col_min)/div

iris
```

```python
def test(x , a,b,c,d):
  # print("----------  " , x[1] , "...............")
  res = int( a*x[1][0]+b*x[1][1]+c*x[1][2]+d*x[1][3] )
  # print( res , " --- " , x[1][4] )
  if( x[1][4] == res ):
    return  0
  return 1 # error


r = 0
for x in iris.iterrows():
  #print( x[1]   ,  " ---- "  ,  x[1][0])
  rx = test(x, -1, 1.5 ,0.4 ,1)
  # -1, 1.5 ,0.4 ,1 -> .80 accuracy
  r += rx

acc = 1-r/len(iris)
print(acc)
```

```python
iris_test_ds = iris.sample(frac = 0.3)
iris_train_ds_index = iris.index.difference(iris_test_ds.index)
iris_train_ds = iris.loc[ iris_train_ds_index ]

```

```python
r = 0
a = 0.2
b = 1.0
c = 0.6
d = 0.2

for x in iris_train_ds.iterrows():
  #print( x[1]   ,  " ---- "  ,  x[1][0])
  rx = test(x, a,b,c,d)
  # -1, 1.5 ,0.4 ,1 -> .80 accuracy
  r += rx

acc =1-r/len(iris_train_ds)
print(acc)
```

```python
r = 0
for x in iris_test_ds.iterrows():
  #print( x[1]   ,  " ---- "  ,  x[1][0])
  rx = test(x, a,b,c,d)
  # -1, 1.5 ,0.4 ,1 -> .80 accuracy
  r += rx

acc = 1-r/len(iris_test_ds)
print(acc)
```

```python
def do_train(a,b,c,d, ds):
  r = 0
  for x in ds.iterrows():
    #print( x[1]   ,  " ---- "  ,  x[1][0])
    rx = test(x, a,b,c,d)
    # -1, 1.5 ,0.4 ,1 -> .80 accuracy
    r += rx


  acc = 1-r/len(ds)
  return acc


def final_test(a,b,c,d,ds):
  r = 0
  for x in ds.iterrows():
    # print( x[1]   ,  " ---- "  ,  x[1][0])
    rx = test(x, a,b,c,d)
    # -1, 1.5 ,0.4 ,1 -> .80 accuracy
    r += rx
    # print(r)

  print(len(ds))
  acc = 1-r/len(ds)
  return acc
```

```python
resultsA = []
resultsB = []
resultsC = []
resultsD = []
resultsE = []

iterations = 0
start = 0
end = 10
for a in range(start,end):
  for b in range(start,end):
    for c in range(start,end):
      for d in range(start,end):
        acc = do_train(a/10,b/10,c/10,d/10, iris_train_ds)
        resultsA.append(a/10)
        resultsB.append(b/10)
        resultsC.append(c/10)
        resultsD.append(d/10)
        resultsE.append(acc)
        iterations += 1
  print( (iterations/100) , "% out of " , 100 , "%")

dct = {"a":resultsA,"b":resultsB,"c":resultsC,"d":resultsD,"acc":resultsE }
results = pd.DataFrame(dct)


```

```python
results.sort_values(by="acc",inplace=True, ascending=False)

params = results.head(1)
sa= float(params["a"])
sb= float(params["b"])
sc= float(params["c"])
sd= float(params["d"])

print(params)
```

```python
acc = final_test(sa,sb,sc,sd,iris_test_ds)
print("Best model acc   " , acc)
```
